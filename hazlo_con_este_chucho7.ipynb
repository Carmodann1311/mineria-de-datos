```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('/content/inViales_2022_2024.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.drop(['folio', 'tipo_incidente_c4', 'incidente_c4'], axis=1)
# Removed the empty string '' from the list of columns to drop.
```

```python
df.head()
```

```python
df = df.drop(['codigo_cierre', 'clas_con_f_alarma', 'tipo_entrada', 'alcaldia_cierre', 'colonia_catalogo', 'longitud', 'latitud'], axis=1)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Crear una instancia del codificador
label_encoder = LabelEncoder()

# Aplicar Label Encoding a la columna tipo_de_entrada y tema_solicitud
df['hora_cierre'] = label_encoder.fit_transform(df['hora_cierre'])
df['dia_semana'] = label_encoder.fit_transform(df['dia_semana'])
```

```python
# Convertir la columna 'fecha' a formato datetime
df['_hora_creacion'] = pd.to_datetime(df['hora_creacion'])
# Filtrar los datos a partir de 2020
df = df[df['_hora_creacion'] >= '00:00:00']
```

```python
# Usar los índices como variable independiente (0, 1, 2, 3)
df['indice'] = df.index

# Definir X e y para la regresión
X = df[['indice']]
y = df['dia_semana']
```

```python
from sklearn.linear_model import LinearRegression

regr = LinearRegression()
```

```python
regr.fit(X,y)
```

```python
pred = regr.predict(X)
```

```python
regr.coef_
```

```python
regr.intercept_
```

```python
print(f"f(x) = x*{regr.coef_[0]} + {regr.intercept_}")
```

```python
import matplotlib.pyplot as plt

# Para tener una mejor visualizacion
fechas = df['hora_creacion']

plt.scatter(fechas, y, color="black")
plt.plot(fechas, pred, color="blue", linewidth=3)
plt.xlabel('hora_creacion')
plt.ylabel('dia_semana')
plt.title('incidentes')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Los valores cercanos al cero indican un mejor ajuste
print('Error cuadrático medio :', mean_squared_error(y, pred))
# El coeficiente de determinacion: 1 es una prediccion perfecta, r > 0 relacion directa y r < 0 relacion inversa.
print("Coeficiente de determinación: %.2f" % r2_score(y, pred))
```
